### Collect AAAI 2024 all paper informations

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait # 브라우저의 응답을 기다릴 수 있게 하기 위해
# 이 외에도 필요한 모듈이 있다면 따로 호출해준다.

import pandas as pd
import time
import random
import json
import os

# ChromeDriver 경로 설정 (다운로드한 ChromeDriver 경로로 대체)
chrome_driver_path = "C:/Users/user_id/your_path_to_chrome_driver/chromedriver.exe"

# Chrome 옵션 설정 (headless 모드로 실행하려면 아래 옵션 활성화)
chrome_options = Options()
chrome_options.add_argument("--headless")  # 브라우저 창을 열지 않고 실행

# Selenium WebDriver 시작 (Chrome driver로 시작)
service = Service(executable_path=chrome_driver_path)
```

```python
driver = webdriver.Chrome(service=service, options=chrome_options)

# AAAI 2024 논문 목록 페이지 URL 설정 (실제 URL로 변경 필요)
aaai_url = "https://ojs.aaai.org/index.php/AAAI/issue/archive" 

# 페이지 열기
driver.get(aaai_url)

# 페이지 로딩을 위한 대기 시간 추가 (필요에 따라 조정 가능)
time.sleep(3)

# 전체 track의 url과 이름을 return
track_urls = []
# 논문 목록의 제목, 초록, 키워드를 탐색 (구조는 실제 웹사이트 구조에 맞춰 수정 필요)
paper_elements = driver.find_elements(By.CSS_SELECTOR, "a.title")
for t in paper_elements:
    track_urls.append(t.get_attribute("href"))
track_urls = track_urls[:-6]
driver.quit()
```

```python
track_urls # AAAI 모든 track 에 접근할 수 있는 url lists
```

```python
topics = {}

for track_site in track_urls:
    print('url: ', track_site)
    # restart driver
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.get(track_site)

    sections = driver.find_elements(By.CSS_SELECTOR, "div.section")


    # 2. 각 section 내의 h2와 모든 a 태그 정보 추출
    for index, section in enumerate(sections):        
        # 해당 section의 h2 텍스트 추출
        h2_text = section.find_element(By.CSS_SELECTOR, "h2").text
        if h2_text not in topics.keys(): 
            topics[h2_text] = []

        
        print(f"Track: {h2_text}")
        
        # 해당 section 내 모든 a 태그의 href와 id 추출
        a_elements = section.find_elements(By.CSS_SELECTOR, "a")
        
        for a in a_elements:
            href = a.get_attribute("href")
            a_id = a.get_attribute("id")
            if len(a_id) > 0:
                topics[h2_text].append([a.text, href])
                print(f"  href: {href}, id: {a_id}", len(a_id))    
    driver.quit()
```

```python
def submission2note(p_title, p_link, track_name):
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.get(p_link)
    
    meta_id = driver.find_element(By.CSS_SELECTOR, 'meta[name="DC.Identifier"]')
    meta_abs = driver.find_element(By.CSS_SELECTOR, 'meta[name="DC.Description"]')
    meta_pdf = driver.find_element(By.CSS_SELECTOR, 'meta[name="citation_pdf_url"]')
    meta_keywords = driver.find_elements(By.CSS_SELECTOR, 'meta[name="citation_keywords"]')

    # Extract the content of the meta tag
    id = meta_id.get_attribute('content')
    abstract = meta_abs.get_attribute('content')
    pdf_link = meta_pdf.get_attribute('content')
    keywords = [i.get_attribute('content') for i in meta_keywords]
    driver.quit()
    
    note = {
        "id": id,
        "title": p_title,
        "keywords": keywords,
        "abstract": abstract,
        "primary_area": track_name,
        "url": pdf_link
    }
    return note
```

```python
# Check the number of total paper
cnt = 0
for i, track_name in enumerate(topics.keys()):
    if i == 0: 
        continue
    cnt += len(topics[track_name])
    notes = [submission2note(p_title, p_link, track_name) for p_title, p_link in topics[track_name]]
    notes = pd.DataFrame(notes)
    print(i, track_name, len(notes))
    notes.to_csv(f'{track_name}.csv', index=False)
    del notes

cnt
```
